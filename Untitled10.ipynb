```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create dataset
data = {
    'Name': ['Asha', 'Ravi', 'Kiran', 'Sneha', 'Arjun', 'Meena', 'Rahul', 'Divya'],
    'Marks': [78, 45, 90, 35, 60, 88, 25, 70]
}

df = pd.DataFrame(data)
df.to_csv('students.csv', index=False)

# Step 2: Load dataset
data = pd.read_csv('students.csv')

# Step 3: Analysis
avg_marks = data['Marks'].mean()
max_marks = data['Marks'].max()
min_marks = data['Marks'].min()

# Pass/Fail
data['Result'] = data['Marks'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')

pass_count = (data['Result'] == 'Pass').sum()
fail_count = (data['Result'] == 'Fail').sum()

# Step 4: Print results
print("----- Student Performance Analysis -----")
print("Average Marks:", avg_marks)
print("Highest Marks:", max_marks)
print("Lowest Marks:", min_marks)
print("Pass Count:", pass_count)
print("Fail Count:", fail_count)

# Step 5: Save output
with open("result.txt", "w") as f:
    f.write(f"Average Marks: {avg_marks}\n")
    f.write(f"Highest Marks: {max_marks}\n")
    f.write(f"Lowest Marks: {min_marks}\n")
    f.write(f"Pass Count: {pass_count}\n")
    f.write(f"Fail Count: {fail_count}\n")

# Step 6: Graph
data['Result'].value_counts().plot(kind='bar')
plt.title("Pass vs Fail")
plt.xlabel("Result")
plt.ylabel("Count")
plt.show()
```
